```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
seri=random.sample(range(100,1000),6)
```

```python
print(seri)
```

```python
plt.figure()                                # Tamamen boş bir grafik oluşturur.
plt.axis()                                  # Sadece eksen çizgileri olan bir grafik oluşturur.
plt.title("Örnek Çizgi Grafik")             # Başlık atar.
plt.xlabel("Yıl")                           # x eksenine isim verir.
plt.ylabel("Fiyatlar")                      # y eksenine isim verir.

x=[2016, 2017, 2018, 2019, 2020, 2021]
y=[563, 911, 937, 480, 908, 771]
z=y-np.mean(y)

plt.plot(x,y, color="red", marker="s", mec="orange", mfc="black", label="seri1")            
# x ve y'yi grafiğin üstüne alır. "marker=" değerleri işaretler.

plt.plot(x,z, linewidth=5, marker="o", ms=10, linestyle="dotted", label="seri2")
# x ve z'yi grafiğin üstüne alır. "ms=" ile de marker büyüklüğü ayarlanır.


plt.legend()                                # .legend() metodunu tanımlamazsak, label'leri göremeyiz.
plt.grid(axis="y")                          # .drid() metodu grafik arkasına ızgara ekler.
plt.show()                                  # .plot metodu ile alınan değerleri çizdirir.

# linestyle="" dotted(:), dashed(--), dashdot(-.) ve solid(-) değerlerini alabilir. Default "solid"tir.
```
