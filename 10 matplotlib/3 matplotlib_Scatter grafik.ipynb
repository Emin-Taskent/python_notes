```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

calisma_suresi=[1.75, 1.5, 0.5, 3, 3.25, 2]
puan=[17,15,7,19,20,16]

plt.scatter(calisma_suresi,puan, marker="^", s=200, edgecolors="black", color="orange", alpha=0.3)

# edgecolor= kenar rengi
# alpha= şeffaflık derecesi
# s= nokta büyüklüğü

plt.axis([0,4,0,21])
plt.grid() 
plt.show()
```

```python
renkler=np.random.randint(0,100,6)                                      
# colorbar için parametre oluşturduk.


plt.scatter(calisma_suresi, puan, s=150, c=renkler, cmap="Reds", edgecolors="black")        
# renk parametresine yukarıdaki değeri atadık ve "cmap="" ile ısı haritası oluşturuldu.

plt.colorbar()
# harita rengi gösterildi


plt.axis([0,4,0,21])
plt.grid() 
plt.show()
```
