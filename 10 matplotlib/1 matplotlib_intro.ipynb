```python
import matplotlib.pyplot as plt
```

```python
seri1=[1,2,3,4,5]

plt.plot(seri1)
plt.show()

# .plot() metodu ile grafik oluşturulur.
# .show() metodu ile de grafik gösterilir.
```

```python
seri2=[6,7,8,9,10]

plt.plot(seri1,seri2)
plt.show()

# .plot() metodu içerisine verilen ilk argüman x eksenini, ikinci argüman ise y eksenine gider.
```

```python
yıl=[2015,2016,2017,2018,2019,2020,2021]
fiyat=[200,300,400,500,600,700,800]

plt.plot(yıl, fiyat)
plt.show()
```

```python
# x ve y kordinatlarının başlangıç ve bitiş değerlerini .axis() metodu ile belirleyebiliriz.

plt.plot(yıl, fiyat)
plt.axis([2015, 2025, 0, 1000])
plt.show()
```

```python
# .title() metodu ile grafiğe başlık da gönderilebilir.

plt.plot(yıl, fiyat)
plt.axis([2015, 2025, 0, 1000])
plt.title("2015-2020 Fiyat")
plt.show()
```

```python
# ._label() metodu ile x veya y eksenine isim tanımlanabilir.

plt.plot(yıl, fiyat)
plt.axis([2015, 2025, 0, 1000])
plt.title("2015-2020 Fiyat")
plt.xlabel("Yıl")
plt.ylabel("Fiyat")
plt.show()
```

```python
# .plot() metoduna parametre göndererek grafikte değişiklik yapabiliriz.

plt.plot(yıl, fiyat, color="red", linewidth=3)
plt.axis([2015, 2025, 0, 1000])
plt.title("2015-2020 Fiyat")
plt.xlabel("Yıl")
plt.ylabel("Fiyat")
plt.show()
```
