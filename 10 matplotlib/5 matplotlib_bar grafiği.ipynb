```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bundan sonrası genel olarak .figure() metodu üzerinden gidecektir.

fig=plt.figure()
grafik=fig.add_axes([0.1, 0.1, 0.8, 0.6])       
# Oluşturulan fig yapısının oluşturulan boş alan içerisinde nerede konumlanacağı belirlenir. Konum için de 4 parametre girilir. Diktörgenin köşeleri gibi düşünülebilir.

araclar=["Fiat", "Ford", "BMW", "Mercedes"]
fiyat=[100, 300, 500, 900]

grafik.bar(araclar, fiyat)
# Grafik değişkeni üzerinden .bar() metodu ile bar grafiği çizeceğini söylüyoruz. Ayrıca içerisine de (x, y) değişkenlerini gönderiyoruz.

plt.show()
```

```python
fig=plt.figure()
grafik=fig.add_axes([0.1, 0.1, 0.8, 0.6])       

araclar=["Fiat", "Ford", "BMW", "Mercedes"]
fiyat=[100, 300, 500, 900]

grafik.bar(araclar, fiyat, color="Red")                 # Renk parametresi gönderdik.

plt.show()
```

```python
fig=plt.figure()
grafik=fig.add_axes([0.1, 0.1, 0.8, 0.6])       

araclar=["Fiat", "Ford", "BMW", "Mercedes"]
fiyat=[100, 300, 500, 900]

grafik.bar(araclar, fiyat, color="Red", width=0.2)                 # Bar kalınlığı parametresi gönderdik.
plt.show()
```

```python
fig=plt.figure()
grafik=fig.add_axes([0.1, 0.1, 0.8, 0.6])       

araclar=["Fiat", "Ford", "BMW", "Mercedes"]
fiyat=[100, 300, 500, 900]

grafik.bar(araclar, fiyat, color="Red", width=0.2, label="Ortalama Fiyat")                 # Bar kalınlığı parametresi gönderdik.
grafik.set_title("Araç Fiyat Grafiği")                                                     # Başlık gönderdik.
grafik.set_xlabel("Marka")                                                                 # x eksenine isim gönderdik.
grafik.set_xlabel("Fiyat")                                                                 # y eksenine isim gönderdik.
grafik.legend()                                                                            # Girilen label'ın legend'ını gönderdik.


plt.show()
```

```python
# Yan bar grafiği çizdirmek istersek .barh() metodunu kullanmamız gerekir.

fig=plt.figure()
grafik=fig.add_axes([0.1, 0.1, 0.8, 0.6])       

araclar=["Fiat", "Ford", "BMW", "Mercedes"]
fiyat=[100, 300, 500, 900]

grafik.barh(araclar, fiyat, color="Red", height=0.2, label="Ortalama Fiyat")              # Bar kalınlığı parametresi gönderdik. height yapısına dikkat!
grafik.set_title("Araç Fiyat Grafiği")                                                    # Başlık gönderdik.
grafik.set_xlabel("Marka")                                                                # x eksenine isim gönderdik.
grafik.set_xlabel("Fiyat")                                                                # y eksenine isim gönderdik.
grafik.legend()                                                                           # Girilen label'ın legend'ını gönderdik.


plt.show()
```

```python
# Örnek

fig2=plt.figure()
grafik2=fig2.add_axes([0.1, 0.1, 0.8, 0.6])

sınıflar=["A", "B", "C", "D", "E"]

matematik=[90, 55, 75, 45, 35]
turkce=[80, 58, 77, 90, 65]
fizik=[40, 35, 68, 78, 80]

grafik2.bar(sınıflar, matematik, label="Matematik", color="Red")
grafik2.bar(sınıflar, turkce, label="Türkçe", color="Blue")
grafik2.bar(sınıflar, fizik, label="Fizik", color="Orange")

grafik2.set_title("Sınıflara Göre Ortalama Not Dağılımı")
grafik2.set_xlabel("Sınıflar")
grafik2.legend()

plt.show()
```

```python
fig2=plt.figure()
grafik2=fig2.add_axes([0.1, 0.1, 0.8, 0.6])

sınıflar=["A", "B", "C", "D", "E"]

indeks=np.arange(len(sınıflar))                                        # arange yapısı için aşağı bakalım.
a=0.25


matematik=[90, 55, 75, 45, 35]
turkce=[80, 58, 77, 90, 65]
fizik=[40, 35, 68, 78, 80]

grafik2.bar(indeks-a, matematik, label="Matematik", color="Red", width=a)
grafik2.bar(indeks, turkce, label="Türkçe", color="Blue", width=a)
grafik2.bar(indeks+a, fizik, label="Fizik", color="Orange", width=a)

grafik2.set_title("Sınıflara Göre Ortalama Not Dağılımı")
grafik2.set_xlabel("Sınıflar")
grafik2.legend()

grafik2.set_xticks(ticks=indeks, labels=sınıflar)

plt.show()
```
