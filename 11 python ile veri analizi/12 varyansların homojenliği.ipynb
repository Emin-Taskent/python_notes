# Varyansların Homojenliği

```python
from scipy import stats 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
np.random.seed(0) 
x1=stats.norm.rvs(loc=0, scale=1, size=1000) 
x2=stats.norm.rvs(loc=0, scale=5, size=1000) 
x3=stats.norm.rvs(loc=0, scale=10, size=1000) 



sns.kdeplot(x1,label="x1")
sns.kdeplot(x2,label="x2")
sns.kdeplot(x3,label="x3")

plt.legend(loc="upper right")
plt.show()
```

```python
x4=stats.norm.rvs(loc=0, scale=1, size=1000)        # yandaki gibi bir popülasyonumuz olduğunu düşünelim.
x41=np.random.choice(x4, size=750)                  # yandaki formül ile x4 popülasyonu üzerinden rasgele 750 adet vei alınıyor.
x42=np.random.choice(x4, size=250)  

x4var=np.var(x4)
x41var=np.var(x41)
x42var=np.var(x42)
```
