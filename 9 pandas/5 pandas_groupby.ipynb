```python
import pandas as pd

```

```python
veri=pd.DataFrame(pd.read_excel("datasets/eczane.xlsx"))


print(veri)
display(veri)
```

```python
# .groupby() metodunun içine gruplanmasını istediğimiz sütun ismini yazarak çalıştırabiliriz.

grup=veri.groupby("BOLGE")
print(grup)

# print ile yazdırdığımızda obje ismi ile karşılaşırız.
```

```python
# Bunu görebilmek için ise for döngüsü kullanılır;

for i in grup:
    print(i)
```

```python
# sadece tek bir gruba ait verileri görmek istersek, 

grup2=veri.groupby("BOLGE").get_group("GAZİEMİR")

print(grup2)
```
