```python
# Bir başka sheet'e erişebilmek için read.excel() metodunun içerisine bir parametre daha göndermemiz gerekir.


veri_doktor=pd.DataFrame(pd.read_excel("C:/Users/Emin Taşkent/Desktop/hastanenorm.xlsx", sheet_name="Doktor"))

print(veri_doktor)
```

```python
# sheet_name= indeks ile de çalışır. Index "0"dan başladığı için bu örnekte "Doktor" sheet'i 4. indekse karşılık gelmektedir.

veri_doktor_2=pd.DataFrame(pd.read_excel("C:/Users/Emin Taşkent/Desktop/hastanenorm.xlsx", sheet_name=4))

print(veri_doktor_2)
```

```python
# Excel dosyası içerisindeki sheet isimlerini görmek istersek;

veri2=pd.ExcelFile("C:/Users/Emin Taşkent/Desktop/hastanenorm.xlsx")
print(veri2)
```

```python
# Yukarıda görüldüğü gibi excel dosyasının içerisindeki bilgiler bir objeye tanımlanmış oldu. Şimdi bu objenin içerisinden istediğimiz bilgileri alabiliriz.

print(veri2.sheet_names)
```

```python
# Obje üzerinden de yeni bir dataframe de oluşturulabilir. 
## "sheet_name=" ifadesinden sonraistenirse index değeri veya sütun adı da kullanılabilir.


veri3=pd.DataFrame(pd.read_excel(veri2,sheet_name=4))               # Aşağıdaki kodla farklı. Döngü kullanılacaksa aşağıdaki gibi yazılmalı.
print(veri3)

```

```python
# Bütün sheetleri yazdırmak için ise for döngüsü kullanılır;

for i in range(0,len(veri2.sheet_names)):
    veri4=pd.DataFrame(pd.read_excel(veri2,sheet_name=veri2.sheet_names[i]))
    print(veri4)
```
