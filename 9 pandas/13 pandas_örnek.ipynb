```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(pd.read_csv("datasets/turkey_car_market.csv"))
df2=df.replace(to_replace="-", value=np.nan)

# Yukarıda veri setinde "-" ile gösterilen ifadeleri "NaN" ifadelere dönüştürdük.

print(df)
```

```python
print(df2)
display(df2)
```

```python
print(df2.info())
```

```python
print(df2.describe())
```

```python
columns=df2.columns

for i in columns:
    print(i)
    print(df2[i].describe())
    print("\n")

# Her kolon ile alakalı olan genel bilgileri yazdırmak için kullanılır.
```

```python
# 0 km olan araçları bulabilmek için;

print(df2["Km"].min())                      # En düşük km olan aracı bulduk burada
print(df2["Km"].min()==df2["Km"])           # İlk kısım yukarıda açıklanmıştır, ikinci kısımda ise df2'nin Km sütununda değeri "0"a eşit olan 
                                            ## satırları boolean olarak döndürüyor.
```

```python
print(df2[df2["Km"].min()==df2["Km"]])      # Yukarıdaki kodu df2'nin içerisine liste olarak gönderirildiğinde, satırları buluyorum.

# 134 adet 0 km araç buldu.
```

```python
# Sıralama yapmak için;

print(df2.sort_values("Fiyat")) 

## Artan şekilde sıralar, print(df2.sort_values("Fiyat", ascending=False)) şeklinde çalıştırılırsa azalan şekilde sıralar.
```
